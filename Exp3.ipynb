```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('parrot.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img,cmap='gray')
plt.title('Grayscale Image')
plt.show()
```

```python
plt.hist(img.ravel(),256,range=[0,256]);
plt.title('Histogram of the Image')
plt.show()
```

```python
img_equalized=cv2.equalizeHist(img)

plt.hist(img_equalized.ravel(),250, range=[0,256])
plt.title('Equalized histogram of the image')
```

```python
plt.imshow(img_equalized, cmap='gray')
plt.title('Original Image')
plt.show()
```

# Histogram Equalization for Color Images

```python
img=cv2.imread('sample.jpg',cv2.IMREAD_COLOR)
```

```python
hsv_img=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
```

```python
hsv_img[:,:,2]=cv2.equalizeHist(hsv_img[:,:,2])
```

```python
equalized_img=cv2.cvtColor(hsv_img,cv2.COLOR_HSV2BGR)
```

```python
plt.imshow(equalized_img[:,:,::-1]); 
plt.title("Equalized Image")
plt.show()
```

```python
plt.hist(equalized_img.ravel(),256, range=[0,256])
plt.title('Equalized Histogram of the Image')
plt.show()
```

```python
plt.subplot(221) 
plt.imshow(img[:, :, ::-1])
plt.title('Original Color Image')

plt.subplot(222)
plt.imshow(equalized_img[:, :, ::-1])
plt.title('Equalized Image')

plt.subplot(223)
plt.hist(img.ravel(),256,range = [0, 256])
plt.title('Original Image')

plt.subplot(224)
plt.hist(equalized_img.ravel(),256,range = [0, 256])
plt.title('Histogram Equalized')

plt.show()
```

```python
plt.figure(figsize = [15,4])
plt.subplot(121)
plt.hist(img.ravel(),256,range = [0, 256])
plt.title('Original Image')

plt.subplot(122)
plt.hist(equalized_img.ravel(),256,range = [0, 256])
plt.title('Histogram Equalized')
```
